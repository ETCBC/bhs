Import text-fabric

````python
from tf.app import use
````

Load BHSA data

````python
A = use("etcbc/bhsa", hoist=globals())
````

# Approach 1: query and export

**NB**: The TSV should load in Excel without hassle, but it is big: 426590 rows. The encoding is UTF16-LE
(otherwise Excel asks questions).

Write a query for all phrases and within them all words.
Retrieve the features `g_word_utf8` and gloss, but do not filter on them (hence the `*`).

````python
query = """
phrase
  word g_word_utf8* gloss*
"""
````

Run the query.

````python
results = A.search(query)
````

Show a preview of the first 5 phrases.

````python
A.show(results, condensed=True, condenseType="phrase", end=5)
````

Export all results to a TSV file (~/Downloads/results.tsv)

````python
A.export(results)
````

# Approach 2: walk and collect

**NB**: This approach leads to a much smaller, cleaner file, still 253203 rows.
The encoding is ordinary utf8.

We make a list of all phrases, and for each phrase we make a tab-separated line with:

* the node number of the phrase
* its gloss
* its Hebrew text

````python
import os
````

````python
phrasesFile = os.path.expanduser(f"~/Downloads/phrases.tsv")
with open(phrasesFile, "w") as fh:
    for phrase in F.otype.s("phrase"):
        text = T.text(phrase)
        gloss = " ".join(F.gloss.v(w) for w in L.d(phrase, otype="word"))
        fh.write(f"{phrase}\t{gloss}\t{text}\n")
````
